```python
import json
import pandas as pd
import numpy as np

# 读取湖南省县区的所有地址的中心经纬度
import json

# 读取GeoJSON文件
with open('湖南省县区.geojson','r',encoding='utf-8') as f:
    data = json.load(f)
# 提取"name"和"center"字段的值
features = data['features']
id_name=[]
center_location=[]
for feature in features:
    name = feature['properties']['name']
    center = feature['properties']['center']
    center_location.append({'Name':name,'X':center[0], 'Y':center[1]})
print(center_location)
df = pd.DataFrame(center_location)
df.to_csv('湖南省县区中心经纬度.csv',index=False,encoding='utf-8-sig')

```
